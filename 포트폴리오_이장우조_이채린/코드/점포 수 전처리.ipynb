```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 파일 불러오기
df= pd.read_csv("/Users/crlee/pandas/2022_점포수.csv",encoding = 'cp949')
df
```

```python
# 필요한 열만 추출
select_col = ['기준_년_코드','기준_분기_코드', '상권_구분_코드_명','상권_코드_명', '점포_수']
df2 = df[select_col]
df2
```

```python
# 기준_년_코드 > 2022년만 추출
df3 = df2[df2['기준_년_코드'] == 2022]
df3
```

```python
# 상권_구분_코드_명 > 골목상권만 추출
df3 = df3[df3['상권_구분_코드_명'] == "골목상권"]
df3
```

```python
# 기준_년_코드 열 삭제
df4 = df3.drop('기준_년_코드', axis=1)
df4
```

```python
# 상권_코드_명 합치기
result = df4.groupby('상권_코드_명')['점포_수'].sum().reset_index()
result
```

```python
# 분기 별 평균값 구하기
result['점포_수'] = result['점포_수'] / 4
result
```

```python
# 소수점 버림
result['점포_수']=result['점포_수'].apply(lambda x: int(x))
result
```

```python
# 새로운 csv 파일로 저장
result.to_csv("/Users/crlee/pandas/2022_점포수_전처리.csv",encoding = 'cp949', index=False)
```
